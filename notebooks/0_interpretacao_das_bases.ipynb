### Dicionario detalhado dos dataframes (`data_normalizada`)

Fonte principal dos dados: https://www.kaggle.com/datasets/jadsonchagas/agriculture-and-food-security-conab-data/data

#### `armazens_cadastrados.csv`
- `identificacao_armazem`: identificador unico do armazem no cadastro.
- `dsc_especie_armazem`: especie/categoria estrutural do armazem.
- `dsc_tipo_armazem`: tipo operacional do armazem.
- `dsc_tipo_entidade`: tipo de entidade responsavel pelo armazem.
- `dsc_tipo_pessoa`: natureza juridica da pessoa cadastrada (fisica/juridica).
- `nom_municipio`: municipio onde o armazem esta localizado.
- `cod_ibge`: codigo IBGE do municipio.
- `uf`: unidade federativa do armazem.
- `qtd_capacidade_estatica(t)`: capacidade total de estocagem estatica, em toneladas.
- `qtd_capacidade_expedicao(t)`: capacidade de expedicao/escoamento, em toneladas.
- `qtd_capacidade_recepcao(t)`: capacidade de recebimento, em toneladas.
- `latitude`: coordenada geografica de latitude do armazem.
- `longitude`: coordenada geografica de longitude do armazem.
- `nome_armazenador`: nome do armazenador/responsavel operacional.
- `endereco`: endereco cadastrado do armazem.
- `email`: email de contato do armazenador/unidade.

#### `custo_producao.csv`
- `empreendimento`: recorte do tipo de empreendimento/sistema produtivo.
- `ano`: ano de referencia do custo.
- `mes`: mes de referencia do custo.
- `ano_mes`: chave temporal consolidada (ano e mes) 202503 por exemplo.
- `produto`: nome do produto agropecuario.
- `id_produto`: identificador numerico do produto.
- `safra`: safra associada ao registro de custo.
- `uf`: unidade federativa do levantamento.
- `municipio`: municipio de referencia do custo.
- `cod_ibge`: codigo IBGE do municipio.
- `unidade_comercializacao`: unidade usada na comercializacao do produto.
- `vlr_custo_variavel_ha`: custo variavel por hectare.
- `vlr_custo_variavel_unidade`: custo variavel por unidade comercializada.
- `vlr_custo_fixo_ha`: custo fixo por hectare.
- `vlr_custo_fixo_unidade`: custo fixo por unidade comercializada.
- `vlr_renda_fator_ha`: remuneracao/renda de fatores por hectare.
- `vlr_renda_fator_unidade`: remuneracao/renda de fatores por unidade comercializada.

#### `estoque.csv`
- `produto`: nome do produto estocado.
- `id_produto`: identificador numerico do produto.
- `nom_municipio`: municipio da informacao de estoque.
- `cod_ibge`: codigo IBGE do municipio.
- `uf`: unidade federativa.
- `num_ano`: ano de referencia do estoque.
- `num_mes`: mes de referencia do estoque.
- `conta_operacional`: tipo de conta operacional do estoque (categoria contabil).
- `qtd_estoque_kg`: quantidade em estoque, em quilogramas.

#### `exportacao_capacidade_estatica.csv`
- `Ano`: ano de referencia.
- `UF`: unidade federativa.
- `Quantidade (mil t)`: capacidade estatica agregada para exportacao, em mil toneladas.

#### `frete.csv`
- `dsc_fonte`: fonte/origem institucional da observacao de frete.
- `municipio_origem`: municipio de origem da rota.
- `cod_ibge_origem`: codigo IBGE do municipio de origem.
- `uf_origem`: UF de origem.
- `municipio_destino`: municipio de destino da rota.
- `cod_ibge_destino`: codigo IBGE do municipio de destino.
- `uf_destino`: UF de destino.
- `ano`: ano de referencia.
- `mes`: mes de referencia.
- `distancia_km`: distancia da rota em quilometros.
- `valor_frete_tonelada`: custo de frete por tonelada transportada.
- `valor_tonelada_km`: custo por tonelada por quilometro.

#### `leilao.csv`
- `id_edital`: identificador do edital do leilao.
- `num_lote`: numero do lote no edital.
- `num_dco`: numero do DCO (documento/codigo da operacao).
- `dsc_dco`: descricao do DCO.
- `dsc_situacao_dco`: situacao do DCO (ex.: aberto, encerrado, homologado).
- `produto`: produto ofertado/negociado.
- `id_produto`: identificador numerico do produto.
- `dsc_tipo_operacao`: tipo de operacao do leilao.
- `dsc_operacao`: descricao da operacao especifica.
- `ano_edital`: ano do edital.
- `mes_edital`: mes do edital.
- `uf_armazem_origem`: UF do armazem de origem do lote.
- `dsc_unidade_comercializacao`: unidade de comercializacao do produto.
- `qtd_ofertada`: quantidade ofertada no lote.
- `qtd_negociada`: quantidade efetivamente negociada.
- `vlr_operacao_s_icms`: valor da operacao sem ICMS.
- `dsc_unidade_medida_ofertada_negociada`: unidade de medida usada em oferta/negociacao.

#### `levantamento_cafe.csv`
- `ano_agricola`: ano agricola de referencia.
- `safra`: safra correspondente.
- `uf`: unidade federativa.
- `produto`: produto (cafe) e possivel classificacao.
- `id_produto`: identificador numerico do produto.
