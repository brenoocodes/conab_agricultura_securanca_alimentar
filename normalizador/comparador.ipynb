Codigo para comparar dataframe e ver se tem dataframes repetidos

```python
import pandas as pd
from pathlib import Path

base_path=Path("../data_normalizada")
files=sorted(base_path.glob("*.csv"))

dataframes={}
excluidos=[]
mantidos=[]

for file in files:
    df=pd.read_csv(file)
    df=df.sort_index(axis=0).sort_index(axis=1).reset_index(drop=True)
    is_duplicate=False
    for kept_name,kept_df in dataframes.items():
        if df.equals(kept_df):
            excluidos.append(file.name)
            is_duplicate=True
            break
    if not is_duplicate:
        dataframes[file.name]=df
        mantidos.append(file.name)

print("Arquivos mantidos:")
print(mantidos)

print("\nArquivos excluídos por serem iguais:")
print(excluidos)
```
